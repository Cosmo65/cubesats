# Энергетический бюджет радиолинии

В русле разработки системы связи чрезвычайно важно знать, какие значения ОСШ (отношение сигнал-шум) можно ожидать на стороне приемника. Фактически, мы можем рассчитать ОСШ как

$$ \frac{S}{N} = \frac{P_{прд}G_{прд}G_{прм}\eta_{прд}\eta_{прм}L_{доп}L}{kT_{ш}B_{ш}} \qquad (1.1) $$

где $S$ - принятая мощность, $N$ - полная шумовая мощность (соотносится со спектральной плотностью шума  $N_0=kT_{ш}$  и двусторонней дисперсией белого шума  $\sigma^2=\frac{N_0}{2}$ ), $P_{прд}$ - передаваемая мощность, $G_{прд}$ и $G_{прм}$ - коэффициенты усиления антенн (относительно изотроптоного излучателя) на передающей и приемной сторонах соответственно, $\eta_{прд}$ и $\eta_{прм}$ - КПД радиопередающего и радиоприемного трактов соответственно $L$ - погонные затухания, $L_{доп}$ - дополнительные затухания (запас), $k$ - постоянная Больцмана, $T_{ш}$ - эквивалентная шумовая температура и $B_{ш}$ - полоса пропускания шума.
Погонные затухания могут быть оценены по формуле Фриза:

$$ L = 20lg\frac{\lambda}{4\pi d}[дБ] \qquad (1.2) $$

где $\lambda$ - длина электромагнитной (ЭМ) волны (относится к несущей частоте $f_0=\frac{c}{λ}$, $c$ - скорость распространения ЭМ волны) и $d$ - наклонное расстояние от спутника до наземной станции. В соответствие с [1, c.98] полоса пропускания шума $B_{ш}$ может быть приближенно оценен как $\gamma B$ где $B$ - полоса пропускания приемника, а $\gamma$ - есть некоторая константа в диапазоне от  1,002 до 1,57 (зависит от конструкции приемника).

Расчет наклонного расстояния в зависимости от угла элевации  ($\phi$) может быть проветен в соотстветствие с [2].

$$ d = \sqrt{(R_З + h)^2 - R_З^2cos\phi ^2} - R_Зsin\phi \qquad (1.3) $$
где  $R_З$ - радиус Земли.

Эквивалентная шумовая температура не является физической температурой антенны. Она равна температуре резистора, который имел бы одинаковую мощность теплового шума в заданной полосе частот. Этот параметр может быть представлен как:

$$ T_{ш} = T_a+T_e \qquad (1.4) $$
 
где $T_a$ - сумма собственных потерь антенны и шума небесного пространства и $T_e$ - шумовая температура приемника. В дополнение, шумовая температура приемника  может быть вычислена через следующую формулу:

$$ T_e=T_0(F_{сис}−1) \qquad (1.5) $$
 
где $T_0$ равно $290K$ и $F_{сис} = 10^{\frac{NF}{10}}$ - шумовой фактор, который может быть оценен через шумовую картину ($NF$) приемной антенны.

В дополнительные потери входят обычно поляризационные потери, а также потери в атмосфере и газах [3] (справочные данные).

# Link budget: path loss and expected SNR

Before we start considering even different types of modulation it is extremely valuable to know what SNR values can be expected on the receiver side. In fact, we can calculate the SNR as

$$ \frac{S}{N} = \frac{P_tG_tG_r\eta_{t}\eta_{r}L_{add}L}{kT_{noise}B_{noise}} \qquad (1.1) $$

Where $S$ is the received power, $N$ is the total termal noise power (relates to noise spectral density  $N_0=kT_{noise}$  and double-sided white noise variance  $\sigma^2=\frac{N_0}{2}$ ), $P_t$ is the transmitted power, $G_t$ and $G_r$ are the antenna gains (relatively to isotropic antenna) on the transmitter and receiver sides respectively, $\eta_{t}$ и $\eta_{r}$ are feeder gains, $L$ is the path losses, $L_{add}$ is additional losses (margin), $k$ is the Boltzmann constant, $T_{noise}$ is the equivalent noise temperature and $B_{noise}$ is the noise bandwidth.

Path losses can be estimated by Friis formula:

$$ L = 20lg\frac{\lambda}{4\pi d}[dB] \qquad (1.2) $$

where $\lambda$ (relates to the carrier frequency  $f_0=\frac{c}{λ}$, $c$ is the speed of the electromagnetic wave) is the wave length and $d$ is the distance between satellite and the ground station. According to [1, p.98] the noise bandwidth $B_{noise}$ can be estimated roughly as $\gamma B$ where $B$ is the receiver bandwidth and $\gamma$ is the constant from 1.002 to 1.57 that relates to configuration of receiver.

Calculation of the distances in dependance of elevation angles ($\phi$) can be provided according to [2]:

$$ d = \sqrt{(R_E + h)^2 - R_E^2cos\phi ^2} - R_Esin\phi \qquad (1.3) $$
where  $R_E$  is the Earth radius.

Equivalent noise temperature is not the physical temperature of an antenna. It is equal to the temperature of a resistor, which would have the same thermal noise power in the given frequency band. This parameter can be represented as:

$$ T_{noise} = T_a+T_e \qquad (1.4) $$
 
where $T_a$ is the sum of antenna losses and sky noise and $T_e$ is the receiver noise temperature. Additionally, receiver noise temperature can be calculated by following formula:

$$ T_e=T_0(F_{sys}−1) \qquad (1.5) $$
 
where $T_0$ is equal to $290K$ and $F_{sys} = 10^{\frac{NF}{10}}$ is the noise factor which can be estimated by noise figure ($NF$) of receive antenna.

The additional losses usually include polarization losses, as well as losses in the atmosphere and gases [3] (reference data).

# Затенения

Расчет запаса на затенения (медленные замирания, крупномасштабные замирания) также может быть принят во внимание.
В соответствие с логарифмической моделью потерь при распространении [4-5] процесс затенения так же влияет на энергетический бюджете:

$$ P_{прм}[дБ] = P_{прд}[дБ] - L_{\Sigma}[дБ] - \chi \qquad (1.6)$$

где $L_{\Sigma}$ - суммарные потери, а $\chi$ - логнормально распределенное случайное значение. Для более или менее точной оценки бюджета канала будем использовать следующую формулу:

$$ \hat{P}_{прм}[дБ] = P_{прд}[дБ] - L_{\Sigma}[дБ] - Е\{\chi\} \qquad (1.7)$$

где $ E\{∗\} $ обозначает математическое ожидание.


# Shadowing

Calculation of marging for shadowing is also can be added into consideration.
According to log-distance path loss model [4-5] shadowing process influences link budget due to it makes a contribution to losses (logorithmic scale):

$$ P_r[dB] = P_t[dB] - L_{\Sigma}[dB] - \chi \qquad (1.6)$$

where $L_{\Sigma}$ is the sum of losses and $\chi$ is the lognormal distributed random value. For more or less precise estimation of link budget let us use following formula:


$$ \hat{P_r}[dB] = P_t[dB] - L_{\Sigma}[dB] - E\{\chi\} \qquad (1.7)$$

where  $E\{∗\}$  denotes expectation.

# Доплеровские замирания и время когерентности канала

Возможные доплеровские сдвиги частоты можно оценить по известной формуле:

$$ f_Д = f_0\frac{v}{c}cos\phi \qquad (1.8)$$

где $v$ - циклическая скорость спутника.

Доплеровский разброс равен $f_{Дmax} = = f\frac{v}{c}$. Время когерентности можно оценить как [6]:
$$ T_c = \frac{1}{f_{Дmax}} \qquad (1.9) $$


# Doppler shift and coherence time

Possible Doppler shifts can be estimated by well-known formula:

$$ f_D = f_0\frac{v}{c}cos\phi \qquad (1.8)$$

where $v$  is the circular velosity of the satellite.

Doppler spread is equal to $f_{Dmax} = f\frac{v}{c}$. Coherence time can be estimated as [6]:
$$ T_c = \frac{1}{f_{Dmax}} \qquad (1.9) $$



# Заключение

Приведем некоторые параметры резюме:

- Начальная точки: несущая частота $f_0$, высота орбиты $h$;
- Характеристики, зависящие от параметров оборудования (регулируемы): передаваемая мощность $P_{прд}$ , полоса пропускания приемника $B$;
- Исходные данные: коэффициент усиления антенн $G_{прд}$ и $ G_{прм}$ , сумма потерь антенны и шум небесного пространства $T_a$ и шумовая картина $NF$ или шумовая температура $T_{ш}$, статистические параметры затенения.

# Summary

Let us provide some parameters summary:

- Initial point: carrier frequency $f_0$, hight of the orbit $h$;
- Equipment dependent parameters (ajustable): transmitted power  $P_t$ , receiver bandwidth $B$;
- Reference data: antenna gains  $G_t$  and  $G_r$ , sum of antenna losses and sky noise  $T_a$ and noise figure  $NF$ or noise temperature $T_{noise}$, statistical parameters of the shadowing.

```python
class LinkBudget:
    def __init__(self, h, gamma, phi = np.pi*np.array(range(0,181,5))/180):
        self.k = 1.38064852*1e-23 #Bolzman constant
        self.c = 3*1e8
        self.h = h
        self.gamma = gamma
        self.phi = phi
    
    def path_loss(self, f_0):
        R_E = 6356863
        lambd = self.c/f_0
        d = np.sqrt((R_E+self.h)**2 - (R_E**2)*np.cos(self.phi)**2) - R_E*np.sin(self.phi)
        L = lambd**2 /(16*(np.pi**2)*d**2)
        return 20*np.log10(L)
        
    def expected_snr(self, f_0, Pt, Gt, Gr, B, Tnoise, eta_t = 10*np.log10(1), eta_r = 10*np.log10(1), Ladd = 5, lognmu = None, lognsigma = None):
        L = 10**(self.path_loss(f_0)/20)
        eta_t_lin = 10**(eta_t/10)
        eta_r_lin = 10**(eta_r/10)
        Gt_lin = 10**(Gt/10)
        Gr_lin = 10**(Gr/10)
        Pinoise = self.gamma*B
        Lnoise = (self.k*Tnoise*Pinoise)
        EIRP = 10*np.log10(Pt*1e3) + Gt
        snr = 10*np.log10(Pt*Gt_lin*Gr_lin*L*eta_t_lin*eta_r_lin/((10**(Ladd/10))*Lnoise));
        if lognmu != None and lognsigma != None:
            chi = np.random.lognormal(lognmu, lognsigma, 1000000)
            e_chi = np.mean(chi)
            snr = snr - e_chi
        return snr, EIRP
    
    def doppler(self, f_0, v):
        fd = f_0*v/self.c*np.cos(self.phi)
        Tc = 1 / (f_0*v/self.c)
        return fd, Tc*1e6
```

```python
phi = np.pi*np.array(range(0,181,5))/180
l = LinkBudget(750*1e3, 1.1)
L = l.path_loss(2.4*1e9)
L_2 = l.path_loss(433*1e6)
L_3 = l.path_loss(144*1e6)
```

```python
plt.plot(180*phi/np.pi, L, '-o', label='2.4 GHz')
plt.plot(180*phi/np.pi, L_2, '-o', label='433 MHz')
plt.plot(180*phi/np.pi, L_3, '-o', label='144 MHz')
plt.title('Path loss')
plt.xlabel('Elevation angles (degrees)')
plt.ylabel('Path loss (dB)')
plt.legend()
plt.grid()
```

```python
snr, EIRP = l.expected_snr(2.4*1e9, 5, 35, 7.3, 10*1e6, 240, lognmu = 0.13, lognsigma = 1)
snr_2, EIRP_2 = l.expected_snr(433*1e6, 5, 35, 7.3, 10*1e6, 240, lognmu = 0.13, lognsigma = 1)
snr_3, EIRP_3 = l.expected_snr(144*1e6, 5, 35, 7.3, 10*1e6, 240, lognmu = 0.13, lognsigma = 1)
```

```python
plt.plot(180*phi/np.pi, snr, '-o', label='2.4 GHz')
plt.plot(180*phi/np.pi, snr_2, '-o', label='433 MHz')
plt.plot(180*phi/np.pi, snr_3, '-o', label='144 MHz')
plt.title('Expected SNRs')
plt.xlabel('Elevation angles (degrees)')
plt.ylabel('SNR (dB)')
plt.legend()
plt.grid()
```

```python
fd, Tc = l.doppler(2.4*1e9, 7.5*1e3)
fd_2, Tc_2 = l.doppler(433*1e6, 7.5*1e3)
fd_3, Tc_3 = l.doppler(144*1e6, 7.5*1e3)
```

```python
plt.plot(180*phi/np.pi, fd, '-o', label='2.4 GHz')
plt.title('Doppler shifts')
plt.xlabel('Elevation angles (degrees)')
plt.ylabel('Doppler shifts (Hz)')
plt.legend()
plt.grid()
```

```python
plt.plot(180*phi/np.pi, fd_2, '-o', label='433 MHz')
plt.plot(180*phi/np.pi, fd_3, '-o', label='144 MHz')
plt.title('Doppler shifts')
plt.xlabel('Elevation angles (degrees)')
plt.ylabel('Doppler shifts (Hz)')
plt.legend()
plt.grid()
```

```python
print('Coherence time 1: '+str(Tc)+' microseconds')
print('Coherence time 2: '+str(Tc_2)+' microseconds')
print('Coherence time 3: '+str(Tc_3)+' microseconds')
```
