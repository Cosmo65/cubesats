## 1.5. Visibility time

Moreover, if we assume that we consider environment without any influence of the atmosphere to the movement of the satellite, circular velocity can be calculated via the classical way \(fig. 1.5.1\):

$$
v = \sqrt{G\frac{M}{R}} \qquad (1.5.1)
$$

 where $G=6.67 \times 10^{-11} m^3 \times kg^{-1} \times s^{-2}$  is the Gravity constant, $M = 6\times 10^{24} kg$  is the Earth mass and $R = R_E + h$  is the orbit radius \(Earth radius plus orbit heigh\).

![Figure 2.8. Cercular velosity of LEO satellites in dependance of the orbit heigh.](https://raw.githubusercontent.com/kirlf/cubesats/master/.gitbook/assets/velosity.png)
*Figure 1.5.1. Cercular velosity of LEO satellites in dependance of the orbit heigh.*

 Visibility time can be calculated based on knowledge about circular velocity and heigh of the orbit. For trajectory that is shown in [fig. 1.4.2](https://github.com/kirlf/cubesats/blob/master/LinkBudget/SatDist.ipynb) this can be done via [\[1\]](https://ieeexplore.ieee.org/document/7506756/):

$$
t=\frac{2(R_E+h)arccos(\frac{R_E}{R_E+h})}{v}
\qquad (1.5.2)
$$

More precise calculation can be done by \[2, p.121\]:

$$
t = \frac{P}{180^0}\arccos \left( \frac{\cos \lambda_{max}}{\cos \lambda_{min}} \right) \qquad (1.5.3)
$$

where $ P = 1.658669 \times 10^{-4} \times  (6378.14 + h[km])^{3/2} $  is the orbit period \(in minutes\).
